```python
from langchain.document_loaders import TextLoader
from langchain.vectorstores import FAISS
with open("/content/drive/MyDrive/Updated_major_dataset.txt" , "r" , encoding="utf8") as f:
  data=f.read()
```

```python
loader=TextLoader("/content/drive/MyDrive/Updated_major_dataset.txt" , encoding="utf8")
```

```python
document=loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500 , chunk_overlap=50)
text_chunks=text_splitter.split_documents(document)
text_chunks
```

```python
print(text_chunks[3].page_content)
```

```python
from langchain.vectorstores import FAISS
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-mpnet-base-v2')
# Extract the page_content from each Document object
text_contents = [doc.page_content for doc in text_chunks]
embeddings = model.encode(text_contents, show_progress_bar=True)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
model_name = "all-mpnet-base-v2"
model_kwargs = {'device': 'cpu'} # Use 'cuda' if you have a GPU
encode_kwargs = {'normalize_embeddings': False}
embeddings = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
vectorstore = FAISS.from_documents(text_chunks, embeddings)
```

```python
retriever=vectorstore.as_retriever()
```

```python
from langchain.prompts import ChatPromptTemplate
```

```python
template="""You are an assistant for question-answering tasks.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know.
Use ten sentences maximum and keep the answer concise.
Question: {question}
Context: {context}
Answer:
"""
```

```python
prompt=ChatPromptTemplate.from_template(template)
```
